```python
import pandas as pd
```

```python
data = {
        "category" : ['action','action','action','action','action','action',
                     'comedy','comedy','comedy','comedy','comedy','comedy',
                     'biopic','biopic','biopic','biopic','biopic','biopic'],
    
        "movie" : ['Baahubali','Avengers','Terminator','Jumanji','Baaghi',
                    'Dhoom', 'Dhamaal','Hera Pheri', 'Golmaal','fukrey','Garam Masala',
                   '3 idiots','Sultan','Dangal','MS Dhoni', 'Bhag Milkha Bhag','Mary Kom',
                   'Sanju'],
    
        "user_1" : [5,4,4.5,5,5,4,4,3,None,4,4,5,4,3,None,4,5,None],
        "user_2" : [None,5,4,5,4,4,3,3,4,3,3,None,4,4,3.5,4.5,3,2.5]
    }
```

```python
dataset = pd.DataFrame(data)
```

```python
dataset
```

```python
temp_df = dataset[(dataset['user_1'] >= 4) & (dataset['user_2'] >= 4)]
```

```python
temp_df
```

```python
countCategory = pd.value_counts(temp_df['category'])
```

```python
cat = countCategory.argmax(countCategory.max())
```

```python
mostCommonMovies = dataset[dataset['category'] == cat]
```
